```python
import geopandas as gpd
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import folium
import branca.colormap as cm
```

# Getting Brazil Data

```python
bra_municipios = gpd.read_file("zip://BR_Municipios_2021.zip")

simplified_bra_municipios = bra_municipios.copy()
simplified_bra_municipios['geometry'] = simplified_bra_municipios['geometry'].simplify(tolerance = 0.01)

simplified_bra_municipios.head()
```

```python
simplified_bra_municipios['CD_MUN'] = simplified_bra_municipios['CD_MUN'].astype(str)
```

```python
bra_castana = pd.read_csv("Brazil Castana Production 2019_2023.csv")
bra_castana['CD_MUN'] = bra_castana['CD_MUN'].astype(str)
bra_castana =bra_castana.fillna(0)
bra_castana.head()
```

```python
bra_soy = pd.read_csv("Brazil Soy Production 2019_2023.csv")
bra_soy['CD_MUN'] = bra_soy['CD_MUN'].astype(str)
bra_soy =bra_soy.fillna(0)
bra_soy.head()
```

```python
bra_gdf = simplified_bra_municipios.merge(bra_castana, how = 'left', left_on = 'CD_MUN', right_on = 'CD_MUN').merge(bra_soy, how = 'left', left_on = 'CD_MUN', right_on = 'CD_MUN')
bra_gdf = bra_gdf.fillna(0)
bra_gdf.head()
```

# Getting Bolivia Data

```python
bol_municipios = gpd.read_file("zip://bolivia_municipios_2015.zip", encoding = "latin-1")

simplified_bol_municipios = bol_municipios.copy()
simplified_bol_municipios['geometry'] = simplified_bol_municipios['geometry'].simplify(tolerance = 0.01)

simplified_bol_municipios.fillna('None', inplace = True)

simplified_bol_municipios = simplified_bol_municipios[simplified_bol_municipios['municipio'] != 'None'].reset_index(drop = True)

# some manual replacements
replacements = {
    "Puerto Menor de Rurrenabaque" : "Rurrenabaque",
    "Santa Ana" : "Santa Ana de Yacuma",
    "Villa Vaca Guzmán" : "Villa Vaca Guzmán (Muyupampa)",
    "Charagua" : "Charagua Iyambae"
}

simplified_bol_municipios['municipio'] = simplified_bol_municipios['municipio'].replace(replacements)

simplified_bol_municipios.loc[(simplified_bol_municipios['departamen'] == "Beni") & (simplified_bol_municipios['municipio'] == "San Javier"), 'municipio'] = "San Javier (BE)"
simplified_bol_municipios.loc[(simplified_bol_municipios['departamen'] == "Beni") & (simplified_bol_municipios['municipio'] == "San Ramón"), 'municipio'] = "San Ramón (BE)"
simplified_bol_municipios.loc[(simplified_bol_municipios['departamen'] == "Cochabamba") & (simplified_bol_municipios['municipio'] == "Entre Ríos"), 'municipio'] = "Entre Ríos (Bulo Bulo)"
simplified_bol_municipios.loc[(simplified_bol_municipios['departamen'] == "Pando") & (simplified_bol_municipios['municipio'] == "San Lorenzo"), 'municipio'] = "San Lorenzo (PA)"
simplified_bol_municipios.loc[(simplified_bol_municipios['departamen'] == "Pando") & (simplified_bol_municipios['municipio'] == "San Pedro"), 'municipio'] = "San Pedro (PA)"
simplified_bol_municipios.loc[(simplified_bol_municipios['departamen'] == "Pando") & (simplified_bol_municipios['municipio'] == "Santa Rosa"), 'municipio'] = "Santa Rosa del Abuná"
simplified_bol_municipios.loc[(simplified_bol_municipios['departamen'] == "Santa Cruz") & (simplified_bol_municipios['municipio'] == "Santa Rosa"), 'municipio'] = "Santa Rosa del Sara"
simplified_bol_municipios.loc[(simplified_bol_municipios['departamen'] == "Santa Cruz") & (simplified_bol_municipios['municipio'] == "San Ignacio"), 'municipio'] = "San Ignacio de Velasco"
simplified_bol_municipios.loc[(simplified_bol_municipios['departamen'] == "Santa Cruz") & (simplified_bol_municipios['municipio'] == "El Puente"), 'municipio'] = "El Puente (SC)"
simplified_bol_municipios.loc[(simplified_bol_municipios['departamen'] == "Santa Cruz") & (simplified_bol_municipios['municipio'] == "Gutiérrez"), 'municipio'] = "Kereimba Iyaambae"
simplified_bol_municipios.loc[(simplified_bol_municipios['departamen'] == "Oruro") & (simplified_bol_municipios['municipio'] == "Totora"), 'municipio'] = "Totora (OR)"
simplified_bol_municipios.loc[(simplified_bol_municipios['departamen'] == "Tarija") & (simplified_bol_municipios['municipio'] == "Entre Ríos"), 'municipio'] = "Entre Ríos (TA)"
simplified_bol_municipios.loc[(simplified_bol_municipios['departamen'] == "Potosí") & (simplified_bol_municipios['municipio'] == "San Pedro"), 'municipio'] = "San Pedro (PO)"


simplified_bol_municipios
```

```python
bol_soy_2023 = pd.read_csv("Bolivia Soy Production 2023.csv")
bol_soy_2022 = pd.read_csv("Bolivia Soy Production 2022.csv")
bol_soy_2021 = pd.read_csv("Bolivia Soy Production 2021.csv")
bol_soy_2020 = pd.read_csv("Bolivia Soy Production 2020.csv")
bol_soy_2019 = pd.read_csv("Bolivia Soy Production 2019.csv")

bol_soy_2023['Municipality'] = bol_soy_2023['Municipality'].astype(str)
bol_soy_2022['Municipality'] = bol_soy_2022['Municipality'].astype(str)
bol_soy_2021['Municipality'] = bol_soy_2021['Municipality'].astype(str)
bol_soy_2020['Municipality'] = bol_soy_2020['Municipality'].astype(str)
bol_soy_2019['Municipality'] = bol_soy_2019['Municipality'].astype(str)

bol_soy_2023 = bol_soy_2023.fillna(0)
bol_soy_2022 = bol_soy_2022.fillna(0)
bol_soy_2021 = bol_soy_2021.fillna(0)
bol_soy_2020 = bol_soy_2020.fillna(0)
bol_soy_2019 = bol_soy_2019.fillna(0)


bol_soy_total = bol_soy_2019.merge(bol_soy_2020, on = 'Municipality', how = 'outer').merge(bol_soy_2021, on = 'Municipality', how = 'outer').merge(bol_soy_2022, on = 'Municipality', how = 'outer').merge(bol_soy_2023, on = 'Municipality', how = 'outer')
bol_soy_total.head()
```

```python
bol_soy_total['Municipality'].nunique()
```

```python
#all 52 matched!
bol_gdf = simplified_bol_municipios.merge(bol_soy_total, how = 'left', left_on = 'municipio', right_on = "Municipality")

print(len(bol_gdf[~bol_gdf['Municipality'].isna()]))
bol_gdf.head()
```

# Joining Country gdfs

```python
bra_gdf_selected_cols = bra_gdf[["geometry", "NM_MUN", "Brazil Nut Production 2019", "Brazil Nut Production 2020", "Brazil Nut Production 2021", "Brazil Nut Production 2022", "Brazil Nut Production 2023", "Soy Production 2019", "Soy Production 2020", "Soy Production 2021", "Soy Production 2022", "Soy Production 2023"]]
bra_gdf_to_concat = bra_gdf_selected_cols.rename(columns = {"NM_MUN" : "Municipality"})

bra_gdf_to_concat['Municipality'] = "Brazil: " + bra_gdf_to_concat['Municipality']

bra_gdf_to_concat.head()
```

```python
bol_gdf_selected_cols = bol_gdf[["geometry", "municipio", "Soy Production 2019", "Soy Production 2020", "Soy Production 2021", "Soy Production 2022", "Soy Production 2023"]]
bol_gdf_to_concat = bol_gdf_selected_cols.rename(columns = {"municipio" : "Municipality"})

bol_gdf_to_concat['Municipality'] = "Bolivia: " + bol_gdf_to_concat['Municipality']
bol_gdf_to_concat
```

```python
bol_gdf_to_concat = bol_gdf_to_concat.to_crs(bra_gdf_to_concat.crs)

base_data = gpd.GeoDataFrame(pd.concat([bra_gdf_to_concat, bol_gdf_to_concat], ignore_index = True)).fillna(0)

base_data
```

```python
base_data.to_file("final_map_data.geojson", driver = "geojson")
```

```python
castana_colormap = cm.StepColormap(['#fff5f0', '#FDE6DD', '#fdccb8', '#fc8f6f', '#f44d37', '#c5161b', '#67000d'],
                                   index = [0, 10, 200, 1000, 2000, 3000, 10000])


soy_colormap = cm.StepColormap(['#fff5f0', '#FDE6DD', '#fdccb8', '#fc8f6f', '#f44d37', '#c5161b', '#67000d', '#48010A'],
                                   index = [0, 1000, 50000, 100000, 250000, 500000, 1000000, 10000000])

def get_color(value, crop, castana_colormap = castana_colormap, soy_colormap = soy_colormap):
    if crop == 'Brazil Nut':
        color = castana_colormap(value)
        opacity =  0.01 if value <= 10 else 0.75
        return color, opacity
    if crop == 'Soy':
        color = soy_colormap(value)
        opacity =  0.1 if value < 1000 else 0.75
        return color, opacity
```

```python
def get_data_layer(base_data, crop, year, show = False):

    if crop == 'Brazil Nut':
        name = f'{year} {crop} Production in Brazil'
        crop_data = base_data[base_data['Municipality'].str[:6] == 'Brazil']

    elif crop == 'Soy':
        name = f'{year} {crop} Production in Bolivia and Brazil'
        crop_data = base_data
    
    layer = folium.GeoJson(
    crop_data,
    name = name,
    show = show,
    style_function = lambda feature: {
        'fillColor': get_color(feature['properties'][f'{crop} Production {year}'], crop)[0],
        'color': 'black',
        'weight': 0.5,
        'fillOpacity': get_color(feature['properties'][f'{crop} Production {year}'], crop)[1]
        },
    highlight_function=lambda feature: {
        'fillColor': 'yellow',   # Change color on hover
        'color': 'black',         # Keep border black on hover
        'weight': 1.5,            # Increase weight on hover
        'fillOpacity': get_color(feature['properties'][f'{crop} Production {year}'], crop)[1]
        },
    tooltip=folium.GeoJsonTooltip(
        fields=['Municipality', f'{crop} Production {year}'],  # Replace with the name of your column
        aliases=['Municipality:', f'{year} {crop} Production (metric tons)'],             # Tooltip alias
        localize=True                   # Localize numbers
        )
    )
    return layer
```

```python
m = folium.Map(location=[-3.4653, -62.2159], zoom_start=5)

castana_2019_layer = get_data_layer(base_data, 'Brazil Nut', '2019')
castana_2020_layer = get_data_layer(base_data, 'Brazil Nut', '2020')
castana_2021_layer = get_data_layer(base_data, 'Brazil Nut', '2021')
castana_2022_layer = get_data_layer(base_data, 'Brazil Nut', '2022')
castana_2023_layer = get_data_layer(base_data, 'Brazil Nut', '2023')


soy_2019_layer = get_data_layer(base_data, 'Soy', '2019')
soy_2020_layer = get_data_layer(base_data, 'Soy', '2020')
soy_2021_layer = get_data_layer(base_data, 'Soy', '2021')
soy_2022_layer = get_data_layer(base_data, 'Soy', '2022')
soy_2023_layer = get_data_layer(base_data, 'Soy', '2023', show = True)

# Create a FeatureGroup to hold the points
ngos_layer_group = folium.FeatureGroup(name='Brazil Nut Organizations')
for name, coordinates in locations.items():
    folium.Marker(location=coordinates, popup=name).add_to(ngos_layer_group)



castana_2019_layer.add_to(m)
castana_2020_layer.add_to(m)
castana_2021_layer.add_to(m)
castana_2022_layer.add_to(m)
castana_2023_layer.add_to(m)
soy_2019_layer.add_to(m)
soy_2020_layer.add_to(m)
soy_2021_layer.add_to(m)
soy_2022_layer.add_to(m)
soy_2023_layer.add_to(m)
ngos_layer_group.add_to(m)

folium.LayerControl(collapsed = False).add_to(m)
```
